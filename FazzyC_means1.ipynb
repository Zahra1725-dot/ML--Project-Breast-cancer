```python
import numpy as np
import matplotlib.pyplot as plt
import skfuzzy as fuzz
```

```python
np.random.seed(0)
cluster1 = np.random.normal(loc=[10, 10], scale=2, size=(100,2))

cluster2 = np.random.normal(loc=[15,12], scale=2, size=(100, 2))

X=np.concatenate((cluster1, cluster2))
```

```python
from sre_constants import error
n_clusters= 2
m =2.0
maxiter= 1000
error= 0.005
```

```python
cntr, u, u0, d, jm, p, fpc = fuzz.cmeans(X.T, n_clusters, m, error=error, maxiter=maxiter)
```

```python
cluster_membership= np.argmax(u, axis=0)
```

```python
plt.figure(figsize=(10, 7))
colors = ['green', 'red']
```

```python
for j in range(n_clusters):
  plt.scatter(X[cluster_membership==j, 0],
              X[cluster_membership==j, 1],
              color=colors[j],
              label=f'Cluster{j+1}')
```

```python
#plot the centers

plt.scatter(cntr[0], cntr[1],
            marker='*', s=250, c='black', label='Cluster Centers')

plt.title('Fuzzy C-Means Clustering of Cancer Tissue Data')
plt.xlabel('Spectral Feature 1')
plt.ylabel('Spectral Feature 2')
plt.legend()
plt.grid(True)
plt.show()
```
